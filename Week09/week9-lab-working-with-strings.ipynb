__Your Task__

- In  `Q12_Part_4_TEXT` they ask the tools the participants are using day to day. However as you might observe the answers are not standardized. 

- Find all the answers with 'jupyter' mentioned in it.

- Find all the participant who are Python users.

- Find all R users. 

```python
list1=[df['Q12_Part_4_TEXT']]
```

```python
pattern = re.compile(r'(Jupyter)|(jupyter)|(jupiter)')
df['Mentioned']= [True if pattern.search(str(s)) != None else False for s in df['Q12_Part_4_TEXT']]

df[df['Mentioned']== True]['Q12_Part_4_TEXT']
```

```python
# another way of doing same thing

df[df['Q12_Part_4_TEXT'].str.contains('jupyter|Jupyter|jupiter', na=False)]
```

```python
pattern = re.compile(r'(Python)|(python)')
df['Mentioned']= [True if pattern.search(str(s)) != None else False for s in df['Q12_Part_4_TEXT']]

df[df['Mentioned']== True]['Q12_Part_4_TEXT']
```

```python
pattern = re.compile(r'\W[rR][(\s)|(s)|(S)|\W]')
df['Mentioned']= [True if pattern.search(str(s)) != None else False for s in df['Q12_Part_4_TEXT']]

df[df['Mentioned']== True]['Q12_Part_4_TEXT']
```

__\W[rR][(\s)|(s)|(S)|\W]   This worked on my all test cases__

\W - starts with any non- word character
[rR] - matches a single character in the list rR (case sensitive)
[(\s)|(s)|(S)|\W] - match anything \s (for 'R studio' or 'r studio' ) | 's'(for Rstudio or rstudio) | 'S'(for 'RStudio') |'\W' (for 'R' of 'r' )
