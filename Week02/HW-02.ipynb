## Simulation of fair die

````python
import random 
import matplotlib.pyplot as plt

trial_count = 500

#definig a function to give fair die outcomes 

def fair_die_simulation(trials):
    options = [1,2,3,4,5,6]
    simulations = []
    
    for i in range (trials):
        n= random.choice(options)
        simulations.append(n)
    return simulations


#ploting the graph of outcome

f = plt.figure(figsize=(6,6))
plt.hist(fair_die_simulation(trial_count), color= '#a05195', alpha= 0.8)
plt.title("Visualization of fair die",fontsize =14,fontweight="bold")
plt.show()
````

## Simulation of biased die

````python
#definig a function to give biased die outcomes 

def biased_die_simuations(trials, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15]):
    options = [1,2,3,4,5,6]
    simulations = []
    for i in range (trials):
        n= random.choices(options,weights)
        simulations.extend(n)
    
    return simulations


#ploting the graph of outcomes 

f = plt.figure(figsize=(5,5))
plt.hist(biased_die_simuations(trial_count), color= '#a05195', alpha= 0.8)
plt.title('Visualization of biased die',fontsize =14,fontweight="bold")
plt.show()
````

## Comparison of fair and biased dice means 

````python
#calculating the mean of 500 outcomes of both the di

import numpy as np

biased_mean_list=[]
unbiased_mean_list=[]
for i in range(500):
    biased = biased_die_simuations(trial_count)
    biased_mean_list.append(np.mean(biased))
    fair = fair_die_simulation(trial_count)
    unbiased_mean_list.append(np.mean(fair))
    
f = plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.title('Biased die Mean',fontsize =14,fontweight="bold")
plt.hist(biased_mean_list, bins=30,color='#003f5c',alpha=.6)

plt.subplot(2,2,2)
plt.title('Unbiased die Mean',fontsize =14,fontweight="bold")
plt.hist(unbiased_mean_list, bins=30, color='#ffa600', alpha=.6)

plt.subplot(2,2,(3,4))
plt.hist(biased_mean_list,color='#003f5c',alpha=.5, bins=40)
plt.hist(unbiased_mean_list, color='#ffa600', alpha=.5, bins=40)
plt.title('Comparing Means',fontsize =14,fontweight="bold")
plt.show()
````
