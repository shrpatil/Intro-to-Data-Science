```python
df
```

```python
df= pd.read_csv(data_url, sep=',', names=['Year', 'Name', 'State','Val_1','Val_2'])
df = df.fillna("")
```

```python

df.dtypes
```

```python
df['Val_1'] = df['Val_1'].astype(str)
df['Val_2'] = df['Val_2'].astype(str)
df
```

```python

df['Val_1'] = df['Val_1'].apply(lambda x: x.strip('.0'))
df['Val_2'] = df['Val_2'].apply(lambda x: x.strip('.0'))

df
```

```python


df['Total']= df['Val_1'].astype(str)+ df['Val_2'].astype(str)
df.dtypes
```

```python
df['Total']= pd.to_numeric(df['Total'])
df=df.sort_values('Year', ascending=True)
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

plt.figure(figsize=(15,6))
ax = plt.axes()
ax.set_xticks(np.arange(0, len(df['Year'])+1,5))



ax.plot(df['Year'],df['Total'])
plt.title("Past Game Record",fontsize =16,fontweight="bold")
plt.xlabel('Year',fontsize =16)
plt.ylabel('Points',fontsize =16)
plt.show()

plt.tight_layout()
```
